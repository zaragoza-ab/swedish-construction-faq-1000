Load the Swedish Construction FAQ dataset (503 bilingual Q&A pairs, CC BY 4.0) and explore it in five minutes.

## 1. Load the dataset

Option A — from Hugging Face:

````python
!pip install -q datasets
from datasets import load_dataset

ds = load_dataset('DecDEPO/swedish-construction-faq')
print(ds)
````

Option B — directly from the GitHub raw URL (no dependencies):

````python
import json, urllib.request

url = 'https://raw.githubusercontent.com/zaragoza-ab/swedish-construction-faq-1000/main/data/faq.json'
with urllib.request.urlopen(url) as r:
    faq = json.load(r)

print(f'Loaded {len(faq)} Q&A pairs')
print('Sample:', faq[0])
````

## 2. Explore categories

````python
from collections import Counter

cats = Counter(item.get('category', 'uncategorized') for item in faq)
for cat, n in cats.most_common(10):
    print(f'{n:3d}  {cat}')
````

## 3. Search by keyword

````python
def search(query, limit=5):
    q = query.lower()
    hits = [x for x in faq if q in x.get('question', '').lower() or q in x.get('answer', '').lower()]
    return hits[:limit]

for hit in search('ROT-avdrag', limit=3):
    print('Q:', hit['question'])
    print('A:', hit['answer'][:200], '...')
    print('---')
````

## 4. Alpaca format for instruction tuning

Ready-to-train Alpaca JSONL is shipped in `data/faq-alpaca.jsonl`:

````python
alpaca_url = 'https://raw.githubusercontent.com/zaragoza-ab/swedish-construction-faq-1000/main/data/faq-alpaca.jsonl'
with urllib.request.urlopen(alpaca_url) as r:
    alpaca = [json.loads(line) for line in r.read().decode('utf-8').splitlines() if line.strip()]

print(f'{len(alpaca)} Alpaca rows')
print(json.dumps(alpaca[0], indent=2, ensure_ascii=False)[:500])
````
